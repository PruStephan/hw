```python
import csvparser
import numpy as np
from dt.model import Object
from dt.model import Model

def format(n):
    if n <= 9:
        return "0" + str(n)
    return str(n)

ds_range = range(1, 22)
ds_paths = []
datasets = []


for n in ds_range:
    prfx = '../data/' + format(n)
    df_train = csvparser.read_file(prfx + '_train.csv')
    df_test = csvparser.read_file(prfx + '_test.csv')
    train_set = [Object(o[:-1], o[-1]) for o in df_train.values]
    test_set = [Object(o[:-1], o[-1]) for o in df_test.values]
    print(df_test)
    datasets.append([train_set, test_set])

def prec(model, objects):
    score = 0
    for object in objects:
        actual = model.predict(object)
        if object.clazz == actual:
            score += 1
    return score / len(objects)

def build_and_test_dt(train_set, test_set, max_height):
    model = Model(train_set, max_height)
    model.build_tree()
    return prec(model, test_set), model
```

```python
results = []
max_heights = [2, 3, 5, 8, 10, 15, 20, 30, 35, 40, 50, 70]
i = 0
for pr in datasets:
    i += 1
    r = []
    for max_height in max_heights:
        train_set, test_set = pr
        model = Model(train_set, max_height)
        model.build_tree()
        r.append(prec(model, test_set))
    print(str(i) + "th dataset evaluated")
    results.append(r)
```

**Sort opt results and find ones with highest, lowest and avg opt heights**

```python
optimal_results = []
optimal_map = {}
for i in range(21):
    optimal_results.append([np.argmax(results[i]), i])
    optimal_map[i] = np.argmax(results[i])

optimal_results.sort(key=lambda x: x[0])
highest = optimal_results[len(max_heights) - 1]
lowest = optimal_results[0]
avg = optimal_results[len(max_heights) // 2]
```


**Draw charts**

```python
import matplotlib.pyplot as plt

def draw(heights, precs):
    plt.plot(heights, precs)
    plt.show()

draw(max_heights, results[highest[1]])
draw(max_heights, results[lowest[1]])
draw(max_heights, results[avg[1]])
```

**Create random subsamples and build tree by them**

```python
def acc(predictf, objects):
    score = 0
    for object in objects:
        actual = predictf(object)
        if object.clazz == actual:
            score += 1
    return score / len(objects)

def rfpredict(object, models):
    class_votes = {}
    for model in models:
        clazz = model.predict(object)
        if clazz not in class_votes:
            class_votes[clazz] = 1
        else:
            class_votes[clazz] += 1
    return max(class_votes, key=class_votes.get)
        


def random_forest(times, n, c):
    max_height = max_heights[optimal_map[n]]
    objects, test_set = datasets[n]
    original_model = Model(objects, max_height)
    original_model.build_tree()
    print("c = " + str(c))
    orig_acc = prec(original_model, test_set)
    print("original: " + str(orig_acc))
    rfmodels = []
    for i in range(times):
        copy = []

        for x in objects:
            copy.append(Object(x.features.copy(), x.clazz))

        np.random.shuffle(copy)
        copy = copy[:2 * len(copy) // 3]
        features_n = len(copy[0].features)
        for i in range(int(c * features_n)):
            random_f_n = np.random.randint(features_n)
            for x in copy:
                x.features[random_f_n] = 0
        model = Model(copy, max_height)
        model.build_tree()
        rfmodels.append(model)
    rf_acc = acc(lambda x: rfpredict(x, rfmodels), test_set)
    print("random forrest: " + str(rf_acc))
    print()
    return [orig_acc, rf_acc]


def run_random_forest_on_datasets():
    times = 10
    for n in range(21):
        random_forest(times, n, 0.3333)
        
```

```python
cs = np.linspace(1/3, 2/3, 10)
results = [random_forest(10, 0, c) for c in cs]
orig_results = [x[0] for x in results]
rf_results = [x[1] for x in results]
plt.plot(cs, orig_results, label='orig')
plt.plot(cs, rf_results, label='rf')
plt.show()
```

```python
run_random_forest_on_datasets()
```
